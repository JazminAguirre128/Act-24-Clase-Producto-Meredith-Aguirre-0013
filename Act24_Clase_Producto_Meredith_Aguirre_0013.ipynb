```python
class Producto:
    def __init__(self, nombre, precio, categoria, stock):
        self.nombre = nombre
        self.precio = precio
        self.categoria = categoria
        self.stock = stock

    def calcularDescuento(self, porcentaje=10):
        descuento = (self.precio * porcentaje) / 100
        precio_final = self.precio - descuento
        print(f"El producto {self.nombre} tiene un {porcentaje}% de descuento. Precio final: ${precio_final}")

    def verificarDisponibilidad(self):
        if self.stock > 0:
            print(f"El producto {self.nombre} está disponible. Unidades: {self.stock}")
        else:
            print(f"El producto {self.nombre} está agotado.")

    def actualizarStock(self, cantidad):
        self.stock += cantidad
        print(f"Stock actualizado de {self.nombre}. Nuevo total: {self.stock} unidades.")

# --- Configuración de Productos ---

# Producto 1
print("Configuración del Producto 1")
nom1 = input("Introduce el nombre del producto: ")
pre1 = float(input("Introduce el precio: "))
cat1 = input("Introduce la categoría: ")
stk1 = int(input("Introduce el stock inicial: "))

producto1 = Producto(nom1, pre1, cat1, stk1)

# Producto 2
print("\nConfiguración del Producto 2")
nom2 = input("Introduce el nombre del producto: ")
pre2 = float(input("Introduce el precio: "))
cat2 = input("Introduce la categoría: ")
stk2 = int(input("Introduce el stock inicial: "))

producto2 = Producto(nom2, pre2, cat2, stk2)

# --- Interacción con los objetos ---

# Operaciones Producto 1
print(f"\n--- Operando Producto 1: {producto1.nombre} ---")
producto1.verificarDisponibilidad()
porcentaje_desc = float(input("¿Qué porcentaje de descuento aplicar?: "))
producto1.calcularDescuento(porcentaje_desc)

# Operaciones Producto 2
print(f"\n--- Operando Producto 2: {producto2.nombre} ---")
n_stock = int(input("¿Cuántas unidades nuevas llegaron al almacén?: "))
producto2.actualizarStock(n_stock)
producto2.verificarDisponibilidad()

print("Programa realizado por Meredith Aguirre NL = 2")
```
